```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv('adult.data')
#data.info()
data = data.drop(['39', ' 77516', ' 13', ' 2174',' 0',' 40'], axis=1)
data.head()
```

# Make a new DataFrame from Private values
# x-axis: "Level of Education" 
# y-axis: "Level of Education".count()

```python
Private_df = data[(data[' State-gov'] == ' Private')]
Private_df.columns
```

```python
numeric = Private_df[' Bachelors'].value_counts()
numeric
```

# Looking at the Private sector

```python
# now we can plot the data
plt.title("United States Private-Sector Educational Level in 1996")
numeric.plot(kind="bar", figsize=(8, 6))
plt.xlabel("Level of Education Achieved")
plt.ylabel("Count (per 1,000)")
plt.xticks(rotation=45)
plt.show()
```
